# **Problemas con Pucciniastrum**

```python
import os

# Rutas de las carpetas
images_folder = r'C:\Users\AlxMa\Desktop\hongos\Pucciniastrum\images'
labels_folder = r'C:\Users\AlxMa\Desktop\hongos\Pucciniastrum\labels'

# Listar archivos en ambas carpetas
images_files = os.listdir(images_folder)
labels_files = os.listdir(labels_folder)

# Obtener los nombres de los archivos de labels sin la extensin
labels_basenames = [os.path.splitext(label)[0] for label in labels_files]

# Recorrer los archivos en images y eliminar los que no tengan un archivo correspondiente en labels
for image_file in images_files:
    image_basename = os.path.splitext(image_file)[0]
    if image_basename not in labels_basenames:
        os.remove(os.path.join(images_folder, image_file))
        print(f'Eliminado {image_file} de la carpeta images.')

print("Proceso completado.")
```

```python
# Obtener los nombres de los archivos de images sin la extensin
images_basenames = [os.path.splitext(image)[0] for image in images_files]

# Encontrar archivos de labels que no tienen un archivo correspondiente en images
missing_labels = [label for label in labels_files if os.path.splitext(label)[0] not in images_basenames]

# Mostrar los nombres de los archivos .txt que no tienen correspondencia
if missing_labels:
    print("Archivos .txt en labels que no tienen un archivo correspondiente en images:")
    for label in missing_labels:
        print(label)
else:
    print("Todos los archivos .txt tienen un archivo correspondiente en images.")

# Guardar los nombres de los archivos .txt en un archivo
with open("missing_labels.txt", "w") as file:
    for label in missing_labels:
        file.write(f"{label}\n")

print("Proceso completado. Los nombres de los archivos faltantes se han guardado en 'missing_labels.txt'.")
```

```python
import os

dir = r"C:\Users\AlxMa\Desktop\hongos"

folders = os.listdir(dir)

for folder in folders:
    labels_dir = os.path.join(dir, folder, 'labels')
    images_dir = os.path.join(dir, folder, 'images')
    
    if os.path.isdir(labels_dir) and os.path.isdir(images_dir):
        labels_files = [f.split('.txt')[0] for f in os.listdir(labels_dir) if f.endswith('.txt')]
        images_files = [f.split('.jpg')[0] for f in os.listdir(images_dir) if f.endswith('.jpg')]
        
        labels_set = set(labels_files)
        images_set = set(images_files)
        
        if len(labels_set) != len(images_set):
            print(f'Difference in file counts in folder: {folder}')
            print(f'Number of label files: {len(labels_set)}')
            print(f'Number of image files: {len(images_set)}')
        
        if labels_set != images_set:
            print(f'Difference in file names in folder: {folder}')
            print('Labels not in images:', labels_set - images_set)
            print('Images not in labels:', images_set - labels_set)
        else:
            print(f'All files match in folder: {folder}')
    else:
        print(f'labels or images folder missing in {folder}')
```

```python
for folder in folders:
    labels_dir = os.path.join(dir, folder, 'labels')
    images_dir = os.path.join(dir, folder, 'images')
    
    if os.path.isdir(labels_dir) and os.path.isdir(images_dir):
        labels_files = [f.split('.txt')[0] for f in os.listdir(labels_dir) if f.endswith('.txt')]
        images_files = [f.split('.jpg')[0] for f in os.listdir(images_dir) if f.endswith('.jpg')]
        
        labels_set = set(labels_files)
        images_set = set(images_files)
        
        # Identificar archivos sobrantes en labels y images
        labels_not_in_images = labels_set - images_set
        images_not_in_labels = images_set - labels_set
        
        # Eliminar archivos sobrantes en labels
        for label in labels_not_in_images:
            file_path = os.path.join(labels_dir, label + '.txt')
            if os.path.isfile(file_path):
                os.remove(file_path)
                print(f'Removed {file_path}')
        
        # Eliminar archivos sobrantes en images
        for image in images_not_in_labels:
            file_path = os.path.join(images_dir, image + '.jpg')
            if os.path.isfile(file_path):
                os.remove(file_path)
                print(f'Removed {file_path}')
        
        # Verificar nuevamente despus de la eliminacin
        labels_files = [f.split('.txt')[0] for f in os.listdir(labels_dir) if f.endswith('.txt')]
        images_files = [f.split('.jpg')[0] for f in os.listdir(images_dir) if f.endswith('.jpg')]
        
        labels_set = set(labels_files)
        images_set = set(images_files)
        
        if labels_set != images_set:
            print(f'After removal, difference still found in folder: {folder}')
            print('Labels not in images:', labels_set - images_set)
            print('Images not in labels:', images_set - labels_set)
        else:
            print(f'After removal, all files match in folder: {folder}')
    else:
        print(f'labels or images folder missing in {folder}')
```

```python
# Ordena los folders para asegurarse de que el ndice corresponda a la clase
folders.sort()

for folder_index, folder in enumerate(folders):
    labels_dir = os.path.join(dir, folder, 'labels')
    
    if os.path.isdir(labels_dir):
        # Lista de archivos .txt en el directorio de labels
        label_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]
        
        for label_file in label_files:
            file_path = os.path.join(labels_dir, label_file)
            
            with open(file_path, 'r') as file:
                lines = file.readlines()
            
            # Actualiza la etiqueta de clase en cada lnea
            new_lines = []
            for line in lines:
                parts = line.split()
                if len(parts) > 0:
                    parts[0] = str(folder_index)
                    new_lines.append(' '.join(parts) + '\n')
            
            # Escribe las nuevas lneas en el archivo
            with open(file_path, 'w') as file:
                file.writelines(new_lines)
        
        print(f'Updated labels in folder: {folder}')
    else:
        print(f'labels folder missing in {folder}')
```

```python
dataset_dir = r"C:\Users\AlxMa\Desktop\hongo_yolo\datasets"
```

```python
import shutil
import random

# Crear directorios de train y valid
train_images_dir = os.path.join(dataset_dir, 'train', 'images')
train_labels_dir = os.path.join(dataset_dir, 'train', 'labels')
valid_images_dir = os.path.join(dataset_dir, 'valid', 'images')
valid_labels_dir = os.path.join(dataset_dir, 'valid', 'labels')

os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(valid_images_dir, exist_ok=True)
os.makedirs(valid_labels_dir, exist_ok=True)

# Directorios de imgenes y etiquetas
images_dir = os.path.join(dataset_dir, 'images')
labels_dir = os.path.join(dataset_dir, 'labels')

# Obtener todos los archivos de imgenes y etiquetas
image_files = [f for f in os.listdir(images_dir) if f.endswith('.jpg')]
label_files = [f for f in os.listdir(labels_dir) if f.endswith('.txt')]

# Asegurarse de que hay la misma cantidad de imgenes y etiquetas
assert len(image_files) == len(label_files), "Mismatch between image and label files count"

# Barajar y dividir
data = list(zip(image_files, label_files))
random.shuffle(data)

split_index = int(len(data) * 0.8)
train_data = data[:split_index]
valid_data = data[split_index:]

# Copiar archivos a train
for image_file, label_file in train_data:
    shutil.copy(os.path.join(images_dir, image_file), os.path.join(train_images_dir, image_file))
    shutil.copy(os.path.join(labels_dir, label_file), os.path.join(train_labels_dir, label_file))

# Copiar archivos a valid
for image_file, label_file in valid_data:
    shutil.copy(os.path.join(images_dir, image_file), os.path.join(valid_images_dir, image_file))
    shutil.copy(os.path.join(labels_dir, label_file), os.path.join(valid_labels_dir, label_file))

print(f'Training data: {len(train_data)} images and labels')
print(f'Validation data: {len(valid_data)} images and labels')
```
